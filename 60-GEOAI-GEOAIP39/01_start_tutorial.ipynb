##### Import Libraries

```python
import geopandas as gpd

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

##### Modeling data fetching with 'shp' file

```python
shp_path = '89_data/shp/NGII_SIDO/Z_NGII_N3A_G0010000.shp'
sido_gdf = gpd.read_file(shp_path)
```

```python
sido_gdf.info()
```

```python
sido_gdf.head()
```

##### Map visualization

* with geopandas

```python
sido_gdf.plot(figsize=(10, 10))
```

```python
sido_gdf.crs
```

* with matplotlib

```python
plt.rcParams['figure.figsize'] = (10, 10)
```

```python
sido_gdf.plot(figsize=(10, 10))

value_of_x = plt.gca().get_xticks()
plt.gca().set_xticklabels(['{:.0f}'.format(x) for x in value_of_x])
value_of_y = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in value_of_y])

plt.show()
```
