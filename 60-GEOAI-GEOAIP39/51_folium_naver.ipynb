```python
import folium
import geo_coding_31 as gc
```

```python
# 두 지점 지정
SP_X, SP_Y = 37.557638657147585, 126.97322081586789
EP_X, EP_Y = 37.57442565618067, 126.98542855257266
START, END = (SP_X, SP_Y), (EP_X, EP_Y)
```

```python
optimal_data = gc.get_optimal_path(START, END)
```

```python
optimal_data
```

```python
# 최대 3개까지 가능
optimal_path_trafast = optimal_data['route']['trafast'][0]['path']
optimal_path_tracomfort = optimal_data['route']['tracomfort'][0]['path']
optimal_path_traoptimal = optimal_data['route']['traoptimal'][0]['path']
```

```python
base_map = folium.Map(
    location=[(START[0]+END[0])/2, (START[1]+END[1])/2],
    tiles='CartoDB positron',
    zoom_start=15
)
base_map
```

```python
# 두 구간을 점으로 표시
# for coord in optimal_path:
#     folium.CircleMarker(
#         location=[coord[1], coord[0]],
#         radius=3,
#         color='green',
#         fill=True,
#         fill_color='green'
#     ).add_to(base_map)

# 두 구간을 선으로 표시
# 실시간 빠른길
optimal_path2_trafast = [[b, a] for a, b in optimal_path_trafast]
folium.PolyLine(locations=optimal_path2_trafast, color='green').add_to(base_map)
# 실시간 편한길
optimal_path2_tracomfort = [[b, a] for a, b in optimal_path_tracomfort]
folium.PolyLine(locations=optimal_path2_tracomfort, color='blue').add_to(base_map)
# 실시간 최적
optimal_path2_traoptimal = [[b, a] for a, b in optimal_path_traoptimal]
folium.PolyLine(locations=optimal_path2_traoptimal, color='red').add_to(base_map)
```

```python
folium.Marker(
    location=START,
    popup='Start',
    icon=folium.Icon(color='blue')
).add_to(base_map)
folium.Marker(
    location=END,
    popup='End',
    icon=folium.Icon(color='red')
).add_to(base_map)
```
