```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(1234)
dates_idx = pd.date_range('2022-01-01', periods=365)
data_df = pd.DataFrame(
    np.random.randn(365), 
    index=dates_idx, 
    columns=['value']
)
```

```python
# 이동평균 계산
window_size = 365 // 12     # 30
rolling_df = data_df.rolling(
    window=window_size,
    center=True,
    min_periods=15
).mean()
```

```python
# 시각화
plt.figure(figsize=(10, 6))
plt.plot(data_df.index, data_df["value"], label="Original Data", color="blue", alpha=0.5)
plt.plot(data_df.index, rolling_df, label=f"{window_size}-Day Rolling Mean", color="red")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Rolling Mean Example")
plt.legend()
plt.show()
```
