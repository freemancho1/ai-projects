* 제외조건: 1 <= 전주수 < 11

```python
df_cnt = read_data('병합 전체카운트')
show_cols = ['계약전력', '기별:전주', '기별:전선', '기별:인입선', '설계:전주', '설계:전선', '설계:인입선']
show_commas(df_cnt[show_cols].head(30))
```

```python
bd_cond = (df_cnt['기별:전주'] > 0) & (df_cnt['기별:전주'] < 11)
ds_cond = (df_cnt['설계:전주'] > 0) & (df_cnt['설계:전주'] < 11)

print(f'기별 데이터셋 기준 레코드수: {df_cnt[bd_cond].shape[0]}')
print(f'설계 데이터셋 기준 레코드수: {df_cnt[ds_cond].shape[0]}')
```
